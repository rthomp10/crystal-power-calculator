# Power Calculator

```python
import numpy as np
import matplotlib.pyplot as plt
```

$P=\frac{1}{2}(\omega C_{eq}V)^2*R_S$

```python
def power(f, ceq, v, rs):
    """Returns power dissapated by the crystal"""
    return 0.5 * (2 * np.pi * f * ceq * v)**2 * rs
```

```python
f   = 24e6
ceq = 6e-12
v   = 1.1
rs  = 100
print(f"Power dissapated = {power(f, ceq, v, rs)*1e6}uW")
```

```python
plt.figure(figsize=(10,5))
v = np.arange(500e-3, 2, 1e-3)
plt.plot(v, 1e6*power(f, ceq, v, rs))
plt.ylabel('Power (uW)')
plt.xlabel('Voltage (V)')
```
